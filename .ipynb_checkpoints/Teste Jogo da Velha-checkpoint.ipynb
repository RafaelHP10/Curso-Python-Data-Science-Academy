# Teste de caso com o ChatGPT

Solicitado ao ChatGPT que cria-se um jogo da velha em python, pelo jupter notebook.

Segue abaixo a codificação.

```python
def exibir_tabuleiro(tabuleiro):
    """
    Função para exibir o tabuleiro.
    """
    print("\n")
    for linha in tabuleiro:
        print(" | ".join(linha))
        print("-" * 5)
    print("\n")

def verificar_vitoria(tabuleiro, jogador):
    """
    Verifica se o jogador venceu.
    """
    # Verificar linhas e colunas
    for i in range(3):
        if all([tabuleiro[i][j] == jogador for j in range(3)]) or all([tabuleiro[j][i] == jogador for j in range(3)]):
            return True
    # Verificar diagonais
    if all([tabuleiro[i][i] == jogador for i in range(3)]) or all([tabuleiro[i][2-i] == jogador for i in range(3)]):
        return True
    return False

def jogar_jogo_da_velha():
    """
    Função principal para jogar o jogo da velha.
    """
    tabuleiro = [[" " for _ in range(3)] for _ in range(3)]
    jogadores = ['X', 'O']
    jogador_atual = 0

    while True:
        exibir_tabuleiro(tabuleiro)
        linha = int(input("Digite o número da linha (0, 1 ou 2): "))
        coluna = int(input("Digite o número da coluna (0, 1 ou 2): "))

        # Verificar se a posição está vazia
        if tabuleiro[linha][coluna] == " ":
            tabuleiro[linha][coluna] = jogadores[jogador_atual]

            # Verificar se o jogador venceu
            if verificar_vitoria(tabuleiro, jogadores[jogador_atual]):
                exibir_tabuleiro(tabuleiro)
                print(f"Parabéns! O jogador {jogadores[jogador_atual]} venceu!")
                break

            # Verificar se houve empate
            if all([posicao != " " for linha in tabuleiro for posicao in linha]):
                exibir_tabuleiro(tabuleiro)
                print("O jogo empatou!")
                break

            # Alternar entre os jogadores
            jogador_atual = (jogador_atual + 1) % 2
        else:
            print("Essa posição já está ocupada. Escolha outra.")

# Iniciar o jogo
jogar_jogo_da_velha()
```
